## Single Variable: Numeric

```python
#Default Histogram
numdisplot = sns.displot(tips_df.tip)
```

```python
#Kernel Density Estimates
numdisplot_kde = sns.displot(tips_df.tip, kind="kde")
```

```python
sns.displot(tips_df.tip, kind="kde")
sns.rugplot(tips_df.tip)
```

## Single Categorical Variable

```python
#Using catplot and counting frequency of occurrence of each value of categorical variable
categoricalcountplot = sns.countplot(x=tips_df.day)
```

## Two Numeric Variables

```python
#Jointplot
size_vs_tip = sns.jointplot(x=tips_df['size'],y=tips_df.tip, kind="kde")
```

## Two Numerical and One Categorical

```python
#Two numerical and categorical used for hue
bill_vs_tip = sns.relplot(x=tips_df.total_bill, y=tips_df.tip, hue=tips_df.sex)
```

## Three Numerical Variables

```python
sns.relplot(x=tips_df.total_bill,y=tips_df.tip, size=tips_df['size'])
```

## One Numerical and One Categorical Data

```python
#Catplot with y axis as numerical value
sns.catplot(x='day',y='tip',data=tips_df , kind = "bar")
```

```python
sns.catplot(x='day',y='tip',data=tips_df , kind = "strip")
```

```python
sns.catplot(x='day',y='tip',data=tips_df , kind = "violin")
```

## One Numerical and Two Categorical

```python
sns.catplot(x="day",y="tip",hue="sex",data=tips_df, kind="bar")
```

```python
sns.catplot(x="sex",y="tip",hue="time",kind="bar",data=tips_df)
```
